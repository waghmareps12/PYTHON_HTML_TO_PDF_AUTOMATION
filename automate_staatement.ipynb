```python
import pandas as pd
from jinja2 import Environment, FileSystemLoader
from weasyprint import HTML

# Sample data
data = [
    {"brand": "1-Day Acuvue Moist (1DM)", "baseline_sales": 392591, "q4_quota": 381402, "growth_rate": -2.85},
    {"brand": "1-Day Acuvue Moist for Astigmatism (1MA)", "baseline_sales": 205350, "q4_quota": 225504, "growth_rate": 9.81},
    {"brand": "1-Day Acuvue Moist Multifocal (1MM)", "baseline_sales": 64776, "q4_quota": 68153, "growth_rate": 5.21},
    # Add more rows as needed
]

# Total and strategic qualifier values
total_baseline_sales = 2246751
total_q4_quota = 2580445
total_growth_rate = 14.85
strategic_qualifier = 162941

# Environment setup
env = Environment(loader=FileSystemLoader('.'))
template = env.get_template('template.html')

# Render HTML
html_out = template.render(
    wwid="123456",
    manager="John Doe",
    territory="Territory A",
    data=data,
    total_baseline_sales=total_baseline_sales,
    total_q4_quota=total_q4_quota,
    total_growth_rate=total_growth_rate,
    strategic_qualifier=strategic_qualifier
)

# Save to an HTML file (optional)
with open('output.html', 'w') as file:
    file.write(html_out)

# Convert HTML to PDF using WeasyPrint
HTML(string=html_out).write_pdf("output.pdf")
```

```python
import pandas as pd
from jinja2 import Environment, FileSystemLoader
import pdfkit

# Sample data
data = [
    {"brand": "1-Day Acuvue Moist (1DM)", "baseline_sales": 392591, "q4_quota": 381402, "growth_rate": -2.85},
    {"brand": "1-Day Acuvue Moist for Astigmatism (1MA)", "baseline_sales": 205350, "q4_quota": 225504, "growth_rate": 9.81},
    {"brand": "1-Day Acuvue Moist Multifocal (1MM)", "baseline_sales": 64776, "q4_quota": 68153, "growth_rate": 5.21},
    # Add more rows as needed
]

# Total and strategic qualifier values
total_baseline_sales = 2246751
total_q4_quota = 2580445
total_growth_rate = 14.85
strategic_qualifier = 162941

# Environment setup
env = Environment(loader=FileSystemLoader('.'))
template = env.get_template('template.html')

# Render HTML
html_out = template.render(
    wwid="123456",
    manager="John Doe",
    territory="Territory A",
    data=data,
    total_baseline_sales=total_baseline_sales,
    total_q4_quota=total_q4_quota,
    total_growth_rate=total_growth_rate,
    strategic_qualifier=strategic_qualifier
)

# Save to an HTML file (optional)
with open('output.html', 'w') as file:
    file.write(html_out)

# Convert HTML to PDF
pdfkit.from_string(html_out, 'output.pdf')
```
